```python
plt.hist(b, density=True, bins=200, histtype="step", color="green", label="exponential")
plt.hist(uni, density=True, bins=100, histtype="step", color="blue", label="uniform")
plt.hist(nor, density=True, bins=100, histtype="step", color="red", label="normal")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
```

```python
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
col_df = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
```

```python
pearson_corr = col_df.corr(method='pearson')
```

```python
pearson_corr
```

```python
spearman_corr = col_df.corr(method='spearman')
```

```python
spearman_corr
```

```python
plt.figure(figsize=(9,8))
sns.heatmap(pearson_corr,annot=True)
plt.title('Pearson Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(9,8))
sns.heatmap(spearman_corr,annot=True)
plt.title('Pearson Correlation Heatmap')
plt.show()
```
