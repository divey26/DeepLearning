Question - 04

```python
import matplotlib.pyplot as plt

# Create subplots
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Plot histogram for each feature
features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

for i, feature in enumerate(sf.columns):
    axes[i].hist(sf[feature], bins=30, color='lightyellow', edgecolor='black' )
    axes[i].set_title(f'Histogram of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Question - 05

```python
import seaborn as sns
```

```python
sns.pairplot(sf, diag_kind='hist')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
sns.pairplot(sf, diag_kind='kde', kind='scatter')
plt.suptitle("Pairplot with KDE on Diagonals", y=1.02)
plt.show()
```

```python

g = sns.PairGrid(sf)

g.map_upper(sns.histplot)

g.map_diag(sns.histplot)

g.map_lower(sns.kdeplot)

plt.suptitle("PairGrid: Hist + KDE on Diagonal, KDE on Lower Half", y=1.02)
plt.show()
```
